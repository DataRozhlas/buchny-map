```python
import pandas as pd
```

```python
d = pd.read_excel('./data/Jehly 2017.xlsx', sheetname='celý rok 2017', skiprows=2)
```

```python
d[['ulice', 'celkem']].to_csv('./data/buchny_ul.csv', encoding='utf-8', index=False)
```

```python
u = pd.read_csv('./scratch/brno_ul.csv')
```

```python
ul = {}
for x in d[['ulice', 'celkem']].to_dict(orient='records'):
    ul[x['ulice']] = x['celkem']
```

```python
ul
```

```python
for i in d.ulice.values:
    if i not in u.Nazev.values:
        print(i)
```

```python
def ass(naz):
    if naz in ul:
        return ul[naz]
    else:
        return None
```

```python
u['buchny'] = u.Nazev.apply(lambda x: ass(x))
```

```python
u[u.buchny == u.buchny][['WKT', 'Nazev', 'buchny']].to_csv('scratch/ulice_buchny.csv', encoding='utf-8', index=False)
```
